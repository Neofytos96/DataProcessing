```python
# CELL6 – 4 marks
# Based on the six QS metrics (i.e., predictor variables), write code to build a linear regression model 
# to estimate the “Overall Score” for the UK universities in the world’s top 200.
import scipy
import statsmodels.api as sm


#data set with all the details for UK universities 
uk_general_table = data.loc[data['Location'] == 'United Kingdom']
#data set with UK's academic reputation
ac_rep = uk_general_table['Academic Reputation']
#data set with UK's employer reputation
emp_rep = uk_general_table['Employer Reputation']
#data set with UK's Faculty Student
fac_stu = uk_general_table['Faculty Student']
#data set with UK's Citations per Faculty
cpf = uk_general_table['Citations per Faculty'] 
#data set with UK's International Faculty
int_fac = uk_general_table['International Faculty']
#data set with UK's International Students
int_stu = uk_general_table['International Students']
#data set with UK's Overall Score
ov_sc = uk_general_table['Overall Score']

#create a list with all the datasets of the 6 QS metrics
qsmetrics = [ac_rep, emp_rep, fac_stu, cpf, int_fac, int_stu]

#data set with UK's QS metrics only 
comp = uk_general_table[['Academic Reputation', 'Employer Reputation', 'Faculty Student',
                        'Citations per Faculty','International Faculty', 'International Students' ]]

comp = sm.add_constant(comp)
results = sm.OLS(ov_sc, comp).fit()

#create list to store the slopes 
slopes_new = []
#for loop to find all the slopes from the graphs of overall score against each QS metric
for i in range(1,7):
        slopes_new.append(round(results.params[i],2))

#define the y intercept
yinterceptnew = results.params[0]


#get the correlation closer to 1 and then using the index of it 
#find the y-intecept of the line with that correlation
count=0
regressionList = []
#a for loop to add all the elements for the regression equation
for i in qsmetrics:
    regressionList.append(slopes_new[count])
    regressionList.append('*')
    regressionList.append(i.name)
    regressionList.append('+')
    count+=1
    
#add to the last element of the list the y intercept   
regressionList.insert(len(regressionList),round(yinterceptnew,2))


# – Print the resulting linear equation in the programme, using rounded values with TWO decimal places.
#print the regression list to give me the regression equation
print('Regression Model: ',sep=' ', end=' ',)
for i in regressionList:
    print (i, sep=' ', end=' ', flush=True)

#----------------------------------------------------------------------

#make data set for Canada with national rank 
data_can = data[data['Location'] == 'Canada']
data_can = data_can.reset_index()
data_can.index = np.arange(1, len(data_can)+1)
del data_can['Location']
del data_can['index']

#build the data set to add the predicted overall score
data_can_filtered = data_can.loc[:,['Institution Name', 'Overall Score']]
data_can_filtered['Predicted Overall Score'] = 0



#for loop in canada table to get the values for qs metrics per uni
count_uni = 0
for i in range (0, len(data_can)):
    countSlopes = 0
    #loop in the metrics to multiply them with the appropriate slope each time
    multiplier = 0
    for j in data_can.iloc[i, 1:7].values:
        multResults = []
        multiplier+=j*slopes_new[countSlopes]
        countSlopes+=1
    
    data_can_filtered.iloc[count_uni, data_can_filtered.columns.get_loc('Predicted Overall Score')] = round(multiplier + yinterceptnew,1)
    count_uni+=1


# Use the above model/equation to predict the “Overall Score” for the universities in Canada.

# – Construct and display a tabular data structure (i.e., a DataFrame), 
# where index represents the “national rank” (i.e., 1, 2, 3…); the first column represents the “Institution Name”; 
# the second column represents the “Overall Score”; and the last column represents the “Predicted Overall Score”.

# – Print the mean squared error (MSE) – the average of the squares of the errors – of the model.

#data set with predicted overall score based on reggression model equation
pred_ov_sc = slopes_new[0]*qsmetrics[0]+ slopes_new[1]*qsmetrics[1]+ slopes_new[2]*qsmetrics[2]+slopes_new[3]*qsmetrics[3]+slopes_new[4]*qsmetrics[4]+slopes_new[5]*qsmetrics[5]+yinterceptnew

#data set with real values for overall score
real_ov_sc = ov_sc


sum_squared_errors = 0
#for loop to subtravt real from predicted values and squared the result 
#then find the sum of the squares 
for v in range (len(real_ov_sc)):
        squared_error = (real_ov_sc.values[v] - pred_ov_sc.values[v])**2
        sum_squared_errors += squared_error
        
        
#calculate the MSE
MSE = sum_squared_errors / len(real_ov_sc)
print('\n')
print ('Mean Squared Error: ', MSE)


data_can_filtered
```
